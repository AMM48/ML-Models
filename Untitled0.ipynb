```python
import pandas as pd
import numpy as np
from statsmodels.tsa.arima.model import ARIMA
from sklearn.metrics import mean_absolute_error, mean_squared_error

def forecast_arima_with_differencing(series, order=(3,1,0), forecast_steps=1):
    model = ARIMA(series, order=order)
    model_fit = model.fit()
    forecast = model_fit.forecast(steps=forecast_steps)
    return forecast[0]

def calculate_metrics(observed, predicted):
    mae = mean_absolute_error(observed, predicted)
    mse = mean_squared_error(observed, predicted)
    rmse = np.sqrt(mse)
    return mae, mse, rmse

file_path = './expenses_sar.csv'
df = pd.read_csv(file_path, index_col='date', parse_dates=True)

train_size = int(len(df) * 0.8)
train, test = df[0:train_size], df[train_size:len(df)]

forecasted_next_month = {}
metrics = {}

for category in train.columns:
    forecasted_value = forecast_arima_with_differencing(train[category])
    forecasted_next_month[category] = forecasted_value

    actual_test_values = test[category].values
    predicted_test_values = [forecasted_value] * len(actual_test_values)
    mae, mse, rmse = calculate_metrics(actual_test_values, predicted_test_values)
    metrics[category] = {'MAE': mae, 'MSE': mse, 'RMSE': rmse}

results = {
    'Category': list(train.columns),
    'Forecast': [forecasted_next_month[category] for category in train.columns],
    'MAE': [metrics[category]['MAE'] for category in train.columns],
    'MSE': [metrics[category]['MSE'] for category in train.columns],
    'RMSE': [metrics[category]['RMSE'] for category in train.columns]
}

results_df = pd.DataFrame(results)
results_df.set_index('Category', inplace=True)
results_df

```
